```python
# TASK 1

target_name = "CLEOFE"  # Your Surname
SEED_NUM = "4"         # Last digit of your student ID
redacted_name = ""

for char in target_name:
    if char in "AEIOU":
        redacted_name += SEED_NUM
    else:
        redacted_name += char

print(f"Original Surname: {target_name}")
print(f"Redacted Output: {redacted_name}")
```

```python
# TASK 2

engineer_name = "CLEOFE"
seed = 4
loops = len(engineer_name)

first_letter_ascii = ord(engineer_name[0])
verification_hash = (first_letter_ascii * seed) - loops

print(f"First Letter ASCII: {first_letter_ascii}")
print(f"Verification Hash: {verification_hash}")
```

```python
# TASK 3

correct_password = "ECE_PASSWORD"
attempts = 0

while True:
    user_input = input("Enter password: ")
    attempts += 1
    
    if user_input == correct_password:
        print("Access Granted!")
        break
    else:
        print("Incorrect. System Locked.")

print(f"Attempts Before Success: {attempts}")
```
